```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, zscore

female = np.genfromtxt('nhanes_adult_female_bmx_2020.csv', delimiter=',', skip_header=1)

weights_female = female[:, 0]  
heights_female = female[:, 1]
heights_female_m = heights_female / 100  
bmi_female = weights_female / (heights_female_m ** 2)

female = np.column_stack((female, bmi_female))

z_female = zscore(female, axis=0, nan_policy='omit')

print("Shape of z_female:", z_female.shape)
print("Number of columns in z_female:", z_female.shape[1])

selected_columns = [1, 0, 6, 5, 7]  
selected_data = z_female[:, selected_columns]
pairplot_data = pd.DataFrame(selected_data, columns=['Height', 'Weight', 'Waist Circumference', 'Hip Circumference', 'BMI'])

sns.pairplot(pairplot_data)
plt.suptitle('Pairplot of Standardized Female Body Metrics')
plt.show()

pearson_corr = np.corrcoef(selected_data, rowvar=False)
spearman_corr = spearmanr(selected_data, axis=0)

print("Pearson's correlation coefficients:")
print(pearson_corr)
print("\nSpearman's correlation coefficients:")
print(spearman_corr.correlation)
```

```python
import numpy as np
import matplotlib.pyplot as plt

male = np.genfromtxt('nhanes_adult_male_bmx_2020.csv', delimiter=',', skip_header=1)
female = np.genfromtxt('nhanes_adult_female_bmx_2020.csv', delimiter=',', skip_header=1)

waist_male = male[:, 6]
height_male = male[:, 1]
hip_male = male[:, 5]

waist_female = female[:, 6]
height_female = female[:, 1]
hip_female = female[:, 5]

waist_height_ratio_male = waist_male / height_male
waist_height_ratio_female = waist_female / height_female

waist_hip_ratio_male = waist_male / hip_male
waist_hip_ratio_female = waist_female / hip_female

male = np.column_stack((male, waist_height_ratio_male, waist_hip_ratio_male))
female = np.column_stack((female, waist_height_ratio_female, waist_hip_ratio_female))

plt.figure(figsize=(10, 6))

data = [waist_height_ratio_female, waist_height_ratio_male, waist_hip_ratio_female, waist_hip_ratio_male]
labels = ['Female Waist/Height', 'Male Waist/Height', 'Female Waist/Hip', 'Male Waist/Hip']

plt.boxplot(data, tick_labels=labels)
plt.title('Comparison of Waist-to-Height and Waist-to-Hip Ratios')
plt.ylabel('Ratio')
plt.grid(True)
plt.show()
```

Explanation of the Output

The box-and-whisker plot compares the distributions of the waist-to-height and waist-to-hip ratios for male and female participants. Each box represents the interquartile range (IQR) of the data, with the line inside the box indicating the median. The "whiskers" extend to show the range of the data, excluding outliers, which are plotted as individual points.

    Female Waist/Height vs. Male Waist/Height: The boxplots for the waist-to-height ratio show the distribution for females and males. A higher median or wider IQR in either group suggests a larger or more variable waist-to-height ratio in that group.
    Female Waist/Hip vs. Male Waist/Hip: Similarly, the boxplots for the waist-to-hip ratio compare females and males. Differences in the median and IQR indicate variations in body shape proportions between the genders.

By examining the plots, you can assess which gender tends to have higher or more varied ratios, providing insights into differences in body composition. For example, if the female waist-to-hip ratio is consistently lower than the male waist-to-hip ratio, it may indicate that females typically have a smaller waist relative to their hips compared to males.


11.Advantages and Disadvantages of BMI, Waist-to-Height Ratio, and Waist-to-Hip Ratio
Body Mass Index (BMI)

Advantages:

    Simple and Inexpensive: BMI is easy to calculate using only height and weight, making it a cost-effective method for assessing body fat.
    Widely Used: It is a standard measure used globally, facilitating comparisons across studies and populations.
    Good Indicator of Obesity: BMI correlates well with body fat for most people, providing a general indication of obesity and related health risks.

Disadvantages:

    Does Not Account for Muscle Mass: BMI does not differentiate between muscle and fat, leading to potential misclassification of muscular individuals as overweight or obese.
    Ignores Fat Distribution: It does not consider where fat is distributed in the body, which can be crucial for assessing health risks.
    Variability Across Populations: BMI cutoffs may not be equally applicable to all ethnic groups due to differences in body composition.

Waist-to-Height Ratio (WHtR)

Advantages:

    Reflects Fat Distribution: WHtR directly considers abdominal fat, which is a better predictor of cardiovascular and metabolic risks than overall body fat.
    Simple to Measure: Requires only waist circumference and height, making it easy to measure and understand.
    Better Health Indicator: Studies suggest that WHtR is a better predictor of health risks compared to BMI, especially for identifying central obesity.

Disadvantages:

    Variation with Age: WHtR can vary with age, and the same cutoffs may not be appropriate for all age groups.
    Less Familiar: It is not as widely used or understood as BMI, which may limit its acceptance in clinical and public health settings.

Waist-to-Hip Ratio (WHR)

Advantages:

    Assesses Body Shape: WHR provides insight into body shape and fat distribution, distinguishing between "apple-shaped" (central obesity) and "pear-shaped" (lower body fat) individuals.
    Predicts Health Risks: A higher WHR is strongly associated with higher risks of cardiovascular diseases and type 2 diabetes.
    Useful in Various Populations: WHR is applicable across different ethnic groups and sexes, providing a versatile measure of health risk.

Disadvantages:

    Measurement Variability: Accurate measurement of waist and hip circumferences can be challenging and may vary between measurers.
    Less Commonly Used: It is not as widely utilized as BMI, which can make comparisons across studies or populations more difficult.
    Does Not Reflect Overall Fat: WHR does not account for overall body fat or muscle mass, potentially missing broader aspects of body composition.
